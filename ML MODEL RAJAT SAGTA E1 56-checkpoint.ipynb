```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import yfinance as yf
from datetime import datetime
%matplotlib inline
```

```python
ticker = "GOOG"
today = datetime.today().strftime('%Y-%m-%d')
data = yf.download(ticker, start="2010-01-01", end=today)

data = data[['Close']]
data.head() 
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
look_back = 60 
def create_dataset(data, look_back=60):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i+look_back, 0])
        y.append(data[i+look_back, 0])
    return np.array(X), np.array(y)
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

X_train.shape, X_test.shape
```

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)
predictions_rescaled = np.concatenate((X_test[:, -1, :-1], predictions), axis=1)
predictions_rescaled = scaler.inverse_transform(predictions_rescaled)[:, -1]
y_test_actual = scaler.inverse_transform(
    np.concatenate((X_test[:, -1, :-1], y_test.reshape(-1, 1)), axis=1)
)[:, -1]
mae = mean_absolute_error(y_test_actual, predictions_rescaled)
rmse = np.sqrt(mean_squared_error(y_test_actual, predictions_rescaled))
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data.index[len(data) - len(y_test_actual):], y_test_actual, color='blue', label='Actual Prices')
plt.plot(data.index[len(data) - len(predictions_rescaled):], predictions_rescaled, color='red', label='Predicted Prices')
plt.title(f"{ticker} Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
def predict_next_month(model, last_sequence, scaler, look_back=60, steps=30):
    future_predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(steps):
        input_sequence = np.reshape(current_sequence, (1, look_back, 1))
        predicted_value = model.predict(input_sequence)[0, 0]
        future_predictions.append(predicted_value)
        current_sequence = np.append(current_sequence[1:], predicted_value)
    future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1)).flatten()
    return future_predictions
last_sequence = X_test[-1, :, 0]
predicted_prices = predict_next_month(model, last_sequence, scaler, look_back=look_back, steps=30)
print("Predicted stock prices for the next 30 days:")

```

```python
last_date = data.index[-1]
future_dates = pd.date_range(last_date + pd.Timedelta(days=1), periods=30)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Historical Prices')
plt.plot(future_dates, predicted_prices, label='Predicted Prices', color='red')
plt.title(f"{ticker} Stock Price Prediction - Next 30 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
future_dates = pd.date_range(data.index[-1] + pd.Timedelta(days=1), periods=30)
predicted_prices_df = pd.DataFrame({
    "Date": future_dates,
    "Predicted Price": predicted_prices
})
print("Prices for Next 30 Days")
print(predicted_prices_df)
predicted_prices_df.to_csv("predicted_prices.csv", index=False)
```
